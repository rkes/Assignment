```python
import pandas as pd
import numpy as np
import heapq
import sys
```

```python
max_mnv=10000;
def GetDataMatrix():
    df = pd.read_csv("dataset.csv")
    df_mat = df.to_numpy();
    for i in range(len(df)):
        df_mat[i][0] -=1
        df_mat[i][1] -=1
    return df_mat;

def GetTotalNumberOfNodes(data):
    vertices = set()
    for point in data:
        vertices.add(point[0])
        vertices.add(point[1])
    return len(vertices);

def GenerateAdjGrph(data):
    adjMat={}
    vertixMap= {}
    for point in data:
        if not point[0] in adjMat:
            adjMat[int(point[0])] = set()
        adjMat[int(point[0])].add(int(point[1]))
        vertixMap[int(point[0])]=set()
        vertixMap[int(point[1])]=set()
    return (adjMat,vertixMap)
    
def GenerateGraph(data):
    n =  GetTotalNumberOfNodes(data);
    matrix = np.zeros((n,n),dtype="float_")
    for point in data:
        matrix[int(point[0]),int(point[1])] += point[2] # add all Amount 
        #matrix[int(point[1]),int(point[0])] = matrix[int(point[0]),int(point[1])] # since graph is undirected
    return matrix;

def CleanupGraph(data):
    s = set()
    for point in data:
        if point[0]!=point[1] and point[2] != 0 :
            s.add((point[0], point[1]))
    return s
def GetKNearestNeighbour(graph,n,k=4):
    print("Calculating K Nearest Neighbour")
    knnGrph ={}
    for i in range(n):
        knnGrph[i] = heapq.nlargest(k ,range(len(graph[i])), graph[i].take)
        showCompletedPercentage("KNearest Neighbour ",i,n);
    print("\n Built K Nearest Neighbour Graph")
    return knnGrph

def processKnnGraph(knnGraph,graph):
    processedKnnGraph={}
    for parent,childNodes in knnGraph.items():
        s = set()
        for childNode in childNodes:
            if graph[parent][childNode ] >0:
                s.add(childNode)
        processedKnnGraph[parent]=s
    return processedKnnGraph

def SNN(n,wtGrph,k=4,kt=3):
    knnGrph=GetKNearestNeighbour(wtGrph,n)
    data=processKnnGraph(knnGrph,wtGrph)
    neighbour_dict = {}
    edgeWt = {}
    clusterSet = []
    for u,neighbours in data.items():
        for v in neighbours:
            uNeighs = data[u];
            vNeighs = data[v];
            if u in vNeighs and v in uNeighs:
                sharedNeigh=uNeighs.intersection(vNeighs)
                if(len(sharedNeigh)>kt):
                    edgeWt[u] = len(sharedNeigh)
                    edgeWt[v] = len(sharedNeigh)
                    clusterSet.append(list(uNeighs.union(vNeighs)))
    return clusterSet;

def MututalNN():
    return ""

def mnv(c1,c2,knnGrph):
    mnvVal=0;
    avg=0;
    total=0;
    for u in c1:
        vectU = knnGrph[u]
        for v in c2:
            vectV = knnGrph[v];
            total = total+1;
            try:
                idxV=vectU.index(v);
            except ValueError:
                idxV=max_mnv;
            try:
                idxU=vectV.index(u);
            except ValueError:
                idxU=max_mnv;
            mnvVal = mnvVal + (idxV+idxU) 
    res=(float(mnvVal))/total;
    return res;
                

def showCompletedPercentage(action,i,n):
    sys.stdout.write('\r')
    num = ((float(i)/float(n-1)))*100.0;
    sys.stdout.write(action+"  %d%%" % num+" Completed ")
    sys.stdout.flush()
```
