```python
import numpy as np
import networkx as nx
import heapq
from collections import defaultdict
import math
import operator
```

```python
maxRow = 0
nodes = []
grph = {}
beta=0.85
count=0

trusted_pages_ratio = 0.4
trusted_pages = []
```

```python
with open('data.txt',"r") as data_file:
    for line in data_file:
        lineValue = line.split("\t")
        u = int(lineValue[0])
        v = int(lineValue[1])
        if u >maxRow:
            maxRow = u
        if v > maxRow:
            maxRow = v
        if u  not in nodes:
            nodes.append(u)
        if v  not in nodes:
            nodes.append(v)
        if u  not in grph:
            grph[u]=[v]
        else:
            grph[u].append(v)
            
        
            
```

```python
MTransitionMatrx = np.zeros((maxRow+1,maxRow+1))
InitialTransitionArray = np.zeros(maxRow+1)
```

```python
for parentNode in grph:
    length = len(grph[parentNode])
    fraction = 1/length
    for childNode in grph[parentNode]:
        MTransitionMatrx[childNode][parentNode] = fraction
```

```python
numberOfNodes = len(nodes)
fraction = 1/numberOfNodes
for i in range(maxRow+1):
    if i in nodes:
        InitialTransitionArray[i] = fraction
```

```python
teleport = (1-beta) * InitialTransitionArray # incase of spider trap and dead end issue comes
MTransitionMatrx = beta*MTransitionMatrx
count=0
```

```python
while(1):
    v1 = np.dot(MTransitionMatrx,InitialTransitionArray)+teleport
    if np.array_equal(v1,InitialTransitionArray):
        break;
    else:
        InitialTransitionArray = v1
        count = count+1
```

```python
print("Iteration Required --> "+str(count))
```

```python
page_rank_score = []
```

```python
for i in range(1,len(InitialTransitionArray)):
    if InitialTransitionArray[i]!=0:
        page_rank_score.append([i,InitialTransitionArray[i]])
        
```

```python
sorted_page_rank_score = sorted(page_rank_score, key = operator.itemgetter(1), reverse = True)
print("Seed Calculated ")
```

```python
no_of_trusted_pages = int(trusted_pages_ratio * len(sorted_page_rank_score))
trusted_pages = [page_info[0] for i, page_info in zip(range(0, no_of_trusted_pages), sorted_page_rank_score)]
fraction = 1 / no_of_trusted_pages
```

```python
v = np.zeros(maxRow + 1)
for i in range(1, maxRow + 1):
    if i in trusted_pages:
        v[i] = fraction
```

```python
teleport = (1-beta) * v # incase of spider trap and dead end issue comes
MTransitionMatrx = beta*MTransitionMatrx
count=0
```

```python
while(1):
    v1 = np.dot(MTransitionMatrx, v) + teleport
    if np.array_equal(v1,v):
        break
    else:
        v = v1
        count += 1
```

```python
page_rank_score_after_trustrank = []
for i in range(1, len(v)):
    if v[i] != 0:
        page_rank_score_after_trustrank.append([i, v[i]])

sorted_page_rank_score_after_trustrank = sorted(page_rank_score_after_trustrank, key = operator.itemgetter(1), reverse = True)
```
